```python
import pandas as pd
import datetime as dt
```

### Importando e tratando a tabela

```python
# Peguei uma data expecífica para trabalhar por não ter sempre os dados do dt.date.today(), mas quando passar para o código esse valor deve ser mudado
hoje = dt.date(2023, 5, 3)

# Importando e mostrando a tabela
df_relatorio_reclamacoes = pd.read_excel('Controle do Relatório.xlsx', sheet_name='RECLAMAÇÕES', usecols='A:H')
display(df_relatorio_reclamacoes)
```

### Buscando dados das reclamações de hoje

```python
df_reclamacoes_hoje = df_relatorio_reclamacoes.loc[df_relatorio_reclamacoes['Data do Recebimento'] == str(hoje)]
display(df_reclamacoes_hoje)
```

### Colocando os valores que deverão ser apresentados numa variável

```python
df_reclamacoes_status = df_reclamacoes_hoje['Status'].value_counts()
df_reclamacoes_rpps = df_reclamacoes_hoje['RPPS'].value_counts()


em_andamento = df_reclamacoes_status['EM ANDAMENTO']
conluido = df_reclamacoes_status['CONCLUÍDO']
total_status = em_andamento + conluido
total_clientes = len(df_reclamacoes_rpps)
```
