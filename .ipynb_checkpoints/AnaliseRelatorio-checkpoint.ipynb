```python
import pandas as pd
import datetime as dt
```

### Importando e tratando a tabela

```python
# Peguei uma data expecífica para trabalhar por não ter sempre os dados do dt.date.today(), mas quando passar para o código esse valor deve ser mudado
hoje = dt.date(2023, 4, 27)
# Importando e mostrando a tabela 
df_relatorio_reclamacoes = pd.read_excel('Controle do Relatório.xlsx', sheet_name='RECLAMAÇÕES', usecols='A:H')
display(df_relatorio_reclamacoes)
# display(df_reclamacoes_hoje)
```

### Buscando dados das reclamações de hoje

```python
df_reclamacoes_hoje = df_relatorio_reclamacoes.loc[df_relatorio_reclamacoes['Data do Recebimento'] == str(hoje)]
display(df_reclamacoes_hoje)
```

### Colocando os valores que deverão ser apresentados numa variável 

```python
df_reclamacoes_status = df_reclamacoes_hoje['Status'].value_counts()
df_reclamacoes_rpps = df_reclamacoes_hoje['RPPS'].value_counts()

try: 
    em_andamento = df_reclamacoes_status['EM ANDAMENTO']
except:
    em_andamento = 0

try:
    conluido = df_reclamacoes_status['CONCLUÍDO']
except:
    conluido = 0

total_status = em_andamento + conluido
total_clientes = len(df_reclamacoes_rpps)


# Armazenando dados num dicionário
dict_reclamacoes = {
    'NÃO RESOLVIDAS' : em_andamento,
    'RESOLVIDAS' : conluido ,
    'TOTAL DE RECLAMAÇÕES' :  total_status,
    'TOTAL DE CLIENTES' :  total_clientes
}

print(dict_reclamacoes)
```

### Importando dados do desenvolvimento

```python
df_relatorio_desenvolvimento = pd.read_excel('Controle do Relatório.xlsx', sheet_name='Desenvolvimento', usecols='A:E')
display(df_relatorio_desenvolvimento)
```

### Analisando dados de desenvolvimento

```python
cont_devStatus = df_relatorio_desenvolvimento['Status'].value_counts()

dev_pendentes = cont_devStatus['Pendente']

df_dev_conluido_hoje = df_relatorio_desenvolvimento.loc[df_relatorio_desenvolvimento['Data Conclusão'] == str(hoje)]
dev_concluidos_hoje = len(df_dev_conluido_hoje)

tot_dev = dev_pendentes + dev_concluidos_hoje

dict_desenvolvimento = {
    'PENDENTES' : dev_pendentes,
    'CONCLUÍDOS' : dev_concluidos_hoje,
    'TOTAL' : tot_dev
}

dict_desenvolvimento
```
